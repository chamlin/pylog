```python
%matplotlib inline
plt.rc('figure', figsize=(15, 8))
```

```python
stats = pd.read_csv ('stats-all.csv', sep='\t', parse_dates=['timestamp'])
```

```python
restarts = pd.unique(stats[stats['restart']>0].index.values)
```

```python
stats['mem-anon-mb'].max()
```

```python
fig, axes = plt.subplots(3, 1, sharex=True, sharey=True)
fig.subplots_adjust(bottom=None, top=None, wspace=0, hspace=0)
axes[0].plot (stats[(stats['mem-anon-mb'] > 0) & (stats['source'] == 1)]['mem-anon-mb'], marker='.', linestyle='', label='node 1')
axes[0].set_title('Anon memory')
axes[0].legend()
axes[0].vlines(restarts, 0, stats['mem-anon-mb'].max(), color='red')
axes[0].grid()
axes[0].set_ylabel('MB')

axes[1].plot (stats[(stats['mem-anon-mb'] > 0) & (stats['source'] == 2)]['mem-anon-mb'])
axes[1].vlines(restarts, 0, stats['mem-anon-mb'].max(), color='red')
axes[1].grid()

axes[2].plot (stats[(stats['mem-anon-mb'] > 0) & (stats['source'] == 3)]['mem-anon-mb'])
axes[2].vlines(restarts, 0, stats['mem-anon-mb'].max(), color='red')
axes[2].set_xlabel('timestamp')
axes[2].grid()
```

```python
fig, axes = plt.subplots()
s2 = stats[(stats['mem-anon-mb'] > 0) & (stats['source'] == 1)]
axes.scatter (s2.index.values, s2['mem-anon-mb'])
#axes.set_xticks(axes.get_xticks())
#axes.set_xticklabels(axes.get_xticks(), rotation = 45)
axes.tick_params(axis='x', labelrotation = 45)
```
